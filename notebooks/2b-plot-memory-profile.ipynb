```python
%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd

plt.style.use('ggplot')
```

Plot memory profile for 128 x 128 image model.

```python
df_1 = pd.read_csv('../output/nmf_0_128.dat', delimiter=' ', skiprows=[0], names=['type', 'usage', 'time'])
df_1['time_relative'] = df_1['time'] - df_1['time'].min()
df_1.head()
```

```python
df_2 = pd.read_csv('../output/nmf_1_128.dat', delimiter=' ', skiprows=[0], names=['type', 'usage', 'time'])
df_2['time_relative'] = df_2['time'] - df_2['time'].min()
df_2.head()
```

```python
plt.figure(figsize=(9,6), constrained_layout=True)
plt.plot(df_1['time_relative'], df_1['usage'], label='In-core')
plt.plot(df_2['time_relative'], df_2['usage'], label='Out-of-core')
plt.xlabel('Time (seconds)', fontsize=16)
plt.ylabel('Memory Usage (MB)', fontsize=16)
plt.title('NMF In-core vs Out-of-Core: (128x128)', fontsize=16)
plt.legend(loc='best', fontsize=16)
plt.show()
```
